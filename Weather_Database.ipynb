```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from citipy import citipy
import random
from datetime import datetime
import requests
from config_weather import api_key
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key
```

```python
! jupyter nbextension list
```

```python
lat = np.random.uniform (-90,90,size=2000)
```

```python
long = np.random.uniform (-180,180,size=2000)
```

```python
final_lat_long = zip(lat,long)
final_lat_long = list(final_lat_long)
```

```python
print (type(final_lat_long))
city_list = []

# Use citipy to find nearest cities using coordinate list
for coordinate in final_lat_long:
    city = citipy.nearest_city(coordinate[0],coordinate[1])
    city_name = city.city_name
    country_name = city.country_code
    
    # Find unique cities to add to cities list
    if city_name not in city_list:
        city_list.append(city_name)
    
len(city_list)
```

```python
city_url = url + "&q=" + 'Boston'
city_weather = requests.get(city_url)
city_weather
```

```python
# Create a list for recording the weather data.
city_data = []

# Create counters.
record_count = 1
set_count = 1

# Loop through all the cities in the list.
for i, city in enumerate(city_list):

    # Group cities in sets of 50 for logging purposes.
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
    # Create endpoint URL with each city.
    city_url = url + "&q=" + city.replace(" ","+")

    # Log the URL, record, and set numbers and the city.
    print(f"{city}")
    
    # Add 1 to the record count.
    record_count += 1
    
# Run an API request for each of the cities.
    try:
        # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()
        
        # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_description = city_weather["weather"][0]["description"]
        
        # Convert the date to ISO standard.
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
       
        # Append the city information into the city_data list.
        city_data.append({"City": city.title(),
                         "Lat": city_lat,
                         "Lng": city_lng,
                         "Max Temp": city_max_temp,
                         "Humidity": city_humidity,
                         "Cloudiness": city_clouds,
                         "Wind Speed": city_wind,
                         "Country": city_country,
                          "Weather Description": city_description,
                         "Date": city_date})
        
    # If an error is experience, skip the city.
    except:
        print("No city found by this name")
        pass
```

```python
len(city_data)
```

```python
# Convert the array of dictionaries to a Pandas DataFrame.
final_data_df = pd.DataFrame(city_data)
final_data_df.head()
```

```python
col_order = ["City", "Country", "Date", "Lat","Lng", "Max Temp", "Humidity","Cloudiness","Wind Speed", "Weather Description"]
final_data_df = final_data_df[col_order]
final_data_df.head(10)
```

```python
# Export the city_data DataFrame into a CSV.
final_data_df.to_csv("WeatherPy_Database.csv", index_label="City_ID")
```
